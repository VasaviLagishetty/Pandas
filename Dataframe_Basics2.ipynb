## Filtering DataFrame based on condition

```
import pandas as pd
```

```
url = "https://raw.githubusercontent.com/VasaviLagishetty/Pandas/master/Datasets/employee.csv"
data = pd.read_csv(url, parse_dates = ["Start Date", "Last Login Time"])
data
```

```
data.dtypes
```

```
data['Senior Management'] = data['Senior Management'].astype('bool')
data.dtypes
```

```
part1 = data['Gender'] == 'Female'
part2 = data['Team'] == 'Product'
```

## and operation

```
data[part1 & part2]
```

## or operation

```
data[part1 | part2]
```

```
part3 = data['Salary'] > 100000
```

```
data[(part1 & part3) | part2]
```

## isnull( ) and notnull( )

```
mask1 = data['First Name'].isnull()

data[mask1]
```

```
mask2 = data['Team'].notnull()

data[mask2]
```

## between( ) method

```
data[data['Start Date'].between('1993-01-01','1998-01-01')]
```

```
data[data['Salary'].between(45000,55000)]
```

```
data[data['Last Login Time'].between("9:30AM","5:30PM")]
```
