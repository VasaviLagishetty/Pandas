## Importing Libraries

```
import pandas as pd
```

## Reading CSV file

```
nba_url = "https://raw.githubusercontent.com/VasaviLagishetty/Pandas/master/Datasets/nba.csv"
nba = pd.read_csv(nba_url)   # reading CSV file
```

## DataFrame Details

```
nba   #dataframe
```

```
nba.info()          # description of dataframe
```

```
nba.columns       #list all columns
```

```
nba.dtypes     #list column and its datatype
```

```
nba.index      # start and end index
```

```
nba.shape    # no of rows and column
```

```
nba.head()   # displays first five records
```

```
nba.tail(10)  # displays last 10 records
```

## Retrieving Column Values

```
nba['Name'] # selecting one specific column
```

```
nba[['Name','College']] # selecting specific columns
```

```
print(nba['Team'] == 'Boston Celtics')
nba[nba['Team'] == 'Boston Celtics']    # retriving rows with specific column value
```

## Adding Column

```
nba['sports'] = 'chess'  # Adding new Column
nba.insert(7, column = "League", value = "National sports")       
nba.insert(3, column = "category", value = "Senior")
nba
```

```
nba.at[0,'sports'] = 'cricket' #to change specific value in the dataframe
nba
```

## Deleting Columns

```
del nba['category']  # deleting columns
nba
```

## Basic Operations

```
nba['Age'] = nba['Age'].add(15)
nba['Age'] = nba['Age'] - 10
nba['Salary'] = nba['Salary'].sub(37)
nba['Position'] = nba['Position'] + '$'
nba['Weight in grams'] = nba['Weight'] * 1000
nba
```

## value_counts() method

```
print(nba['Position'].value_counts())      # counts of Each unique values in a column
print(nba['College'].value_counts())
```

## Filling NaN values in DataFrame

```
nba["College"] = nba["College"].fillna('No_College') # filling NaN values for specific column
nba
```

```
nba['Salary'] = nba['Salary'].fillna(nba['Salary'].mean())    # filling NaN in salary with mean of all rows
nba 
```

## Columns Type Conversion using astype( ) method

```
nba['Salary'] = nba['Salary'].astype('int')        # converting datatypes of a specific column
nba
```

## Sorting Values According to Specific Columns

```
nba = nba.sort_values("Name",ascending=False)      # sorting values acc to specific column
nba.head()
```

```
nba = nba.sort_values("College",ascending=True)
nba.tail()
```

```
nba = nba.sort_values(['Age','Name'],ascending=[False,True])
nba
```

## Sorting Based on Index

```
nba = nba.sort_index(ascending=False)
```
