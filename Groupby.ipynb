```
import pandas as pd
```

```
url = "https://raw.githubusercontent.com/VasaviLagishetty/Pandas/master/Datasets/fortune1000.csv"
fortune = pd.read_csv(url,index_col='Rank')
```

```
fortune.head(5)
```

```
sectors = fortune.groupby('Sector')
```

```
type(sectors)
```

```
sectors.size()
```

# Retrieving particular group using .get_group( ) method

```
sectors.get_group("Energy")
```

# Aggregations on groups

```
sectors.get_group("Technology")['Employees'].sum()
```

```
sectors.get_group("Technology")[['Employees','Revenue']].sum()
```

```
sectors.max()
sectors.min()
sectors.sum()
sectors.mean()
```

# groupby multiple columns

```
sector = fortune.groupby(['Sector','Industry'])
```

```
sector.size()
```

```
sector["Employees"].max()
```

# .agg( ) method

```
sector.agg({"Revenue" : ["sum", "mean"],
             "Profits" : "sum",
              "Employees" : "mean"})
```

```
sector.agg(["size", "sum", "mean"])
```

# Iteration in groups

```
df = pd.DataFrame(columns = fortune.columns)
df
```

```
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_company_in_group)
```

```
df
```

```
cities = fortune.groupby("Location")
df = pd.DataFrame(columns = fortune.columns)
df
```

```
for city, data in cities:
    highest_revenue_in_city = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_in_city)
```
